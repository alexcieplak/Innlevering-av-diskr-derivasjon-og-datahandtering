# Numerisk derivasjon oppgaver 
## Oppgave a
Hva er en sekant?
En sekant er en linje, som i motsetning til en tagent, skjærer gjennom to punkter, i stedenfor en.

$$ k = \frac {f(x + \Delta x) - f(x)}{\Delta x} $$ 

Vi kan skrive stigningstallet til k ved funskjonen f på den måten siden delta a er differansen mellom to punkter. HVis vi har en sekant, kan de to punktene befinne seg hvor som helst på grafen, og $\Delta x$ kan vare utrolig stort eller utrlig lite. Jo større tall man heler på, jo mindre nøyaktig vil stigningstallet til funksjonen være.

## Oppgave b 
Hva er en tangent? 
En tangent er en linje som skjærer ett punkt på en graf. Stigningstallet til tangenten til et punkt er den deriverte til det punktet.

$$f' (x) = \lim_{\Delta x \to 0} \frac {f(x + \Delta x) - f(x)}{\Delta x} $$

Vi kan skrive stigningstallet til k ved funksjonen f på den måten siden denne funskjonen innehloder en grenseverdi. Når vi har en tangent, er det nesten det samme som en sekant av to punkter veldig, veldig nærme hverandre. $\Delta x$ går mot 0. Vi deler på et så lite tall som mulig for å få et så nøyaktig svar som mulig.

## Oppgave d
Newtons kvotient bruker variabelen h istedenfor $\Delta x$. Hvis funskjonen ikke har en grenseverdi som gjør at $\Delta x$ går mot 0, vil det ikke være en tangent til punktet, men en sekant av punktet. 

Newtons kvotient bruker h, dermed deler han det på sekanten av to punkt, ikke tangenten.

$$f' (x) =  \frac {f(x + h) - f(x)}{h} $$

```python
def f(x):
    return (x**2 + 2*x)


def fder(f, x, h = 1E-8):
    df = (f(x + h) - f(x)) / (h)
    return df 

df = fder(f,1)
print(df)
```

# Heisforsøk

```python
from pylab import *

data = loadtxt('height.txt', delimiter = ',', skiprows = 1)

tid = data [:,0]
hoyde = data [:,2]


plot(tid, hoyde, color = 'skyblue')
title('veigraf')
xlabel('tid i sekunder')
ylabel('høyde i meter')
show()

v = [0]


for i in range(len(tid)-1):
    fart = (hoyde[i+1]-hoyde[i])/(tid[i + 1]- tid [i])
    v.append(fart)
      
plot(tid,v, color = 'seagreen')
title('fartsgraf')
xlabel('tid i seunder')
ylabel('fart i m/s')
show()

a= [0]

for i in range(len(tid)-1):
    akselerasjon = (v[i+1]-v[i])/(tid[i + 1]- tid [i])
    a.append(akselerasjon)
    
plot(tid, a, color = 'darkblue')
title('akselerasjonsgraf')
xlabel('tid i seunder')
ylabel('akselerasjon i m/s^2')
show()
```
